```python
import torch
import numpy as np
import torch.nn as nn
```

```python
features = torch.rand(50,4)*120
w = torch.rand(1,4,requires_grad=True)
b = torch.rand(1,requires_grad=True)
```

```python
print(features,w,b)
```

```python
def predict(x):
    return x@w.t() + b
```

```python
labels = torch.from_numpy(np.random.rand(50,1) * 120)
```

```python
from torch.utils.data import DataLoader, TensorDataset
```

```python
train_dl = DataLoader(TensorDataset(features,labels),batch_size=4,shuffle=True)
```

```python
for xb,yb in train_dl:
    print(xb,yb)
    break
```

```python
def mse(preds,labels):
    return torch.sum((preds - labels)**2) / torch.numel(labels)
```

```python
def fit():
    global w,b
    print(w,b)
    for i in range(100):
        for xb,yb in train_dl:
            loss = mse(predict(xb),yb)
            loss.backward()
            print(loss)
            with torch.no_grad():
                w = w - w.grad*1e-3
                b = b - b.grad*1e-3
                w.grad.zero_()
                b.grad.zero_()
```

```python
fit()
```
