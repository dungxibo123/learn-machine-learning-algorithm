```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
#constant
dx = 0.001 #small dx

T = 10
x = np.arange(-T + dx,  T + dx, dx)
T *= 2
L = T / 2
n = len(x)
```

```python
def __f(x):
    return x**3 / 10

#test __f
#__f(1)
```

```python
def get_cmap(n, name='hsv'):
    '''Returns a function that maps each index in 0, 1, ..., n-1 to a distinct 
    RGB color; the keyword argument name must be a standard mpl colormap name.'''
    return plt.cm.get_cmap(name, n)
```

```python
f = np.array(list(map(__f,x)))

```

```python
#compute fourier series
A0 = np.sum(f * np.ones_like(x)) * dx * 2 / T
fFs = A0 / 2
number_of_freq = 20
A = np.zeros(number_of_freq)
B = np.zeros(number_of_freq)
cmap = get_cmap(number_of_freq)

fig, axe1 = plt.subplots(figsize=(10,10))
axe1.plot(x,f)

for k in range(number_of_freq):
    A[k] = 2 / T * np.sum(f * np.cos((k + 1) * x * np.pi / L)) * dx #inner product
    B[k] = 2 / T * np.sum(f * np.sin((k + 1) * x * np.pi / L)) * dx
    fFs +=  A[k] * np.cos((k + 1) * x * np.pi / L) + B[k] * np.sin((k + 1) * x * np.pi / L)
    # L = np.pi ~ 3.14 ~ theo steve brunton
    #print(f, fFs,sep='\n')
    #assert fFs.shape == x.shape
    axe1.plot(x,fFs,'-',color=cmap(k))
```
